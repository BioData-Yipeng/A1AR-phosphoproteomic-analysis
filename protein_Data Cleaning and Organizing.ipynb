```python
import pandas as pd
import re
```

```python
#start data file
df_all=pd.read_csv('protein_PerseusNormalizedData.csv')
```

```python
df_all["unique gene names"] = df_all.apply(
    lambda row: row["Gene names"] if pd.notna(row["Gene names"]) else row["Protein IDs"].split(";")[0],
    axis=1
)
```

```python
df_all["unique gene names"].is_unique
```

```python
# Create a new column to annotate duplicates
df_all['dup_count'] = df_all.groupby('unique gene names').cumcount()
df_all["Uniqueprotein"] = df_all.apply(
    lambda row: f"{row['unique gene names']}_iso{row['dup_count']}" if row['dup_count']>0 else row['unique gene names'],
    axis=1
)
df_all.drop(columns='dup_count', inplace=True)
```

```python
#confirm no replicates in the column
df_all["Uniqueprotein"].duplicated().any() 
```

```python
#subset dataframe
df=df_all.set_index('Uniqueprotein').iloc[:,:16]
```

```python
#end data file
df.to_csv("Uniqueprotein.csv")
```
